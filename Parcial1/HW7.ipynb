## <font color='cornflowerblue'> Problema 2

Using Taylor Series and Itô's lemma derive **Model 5** using the diffusion process $dS_t = S_t \sigma dW_t$, find a differential form of a function $f(S_{(t+\Delta t)}, t + \Delta t) = df_{t + \Delta t} := df_t$

Se tiene una función $f(x,y)$, su expansión de Taylor con $k=3$ es la siguiente:

$$
T(x,y) = f(a,b) + \frac{1}{1!} [\frac{\partial f(a,b)}{\partial x} (x-a) + \frac{\partial f(a,b)}{\partial y} (y-b)] + \\

\frac{1}{2!} [\frac{\partial^2 f(a,b)}{\partial x^2} (x-a)^2 + \frac{\partial^2 f(a,b)}{\partial y^2} (y-b)^2 + 2 \frac{\partial^2 f(a,b)}{\partial x \partial y} (x-a)(y-b)] + \\

\frac{1}{3!} [\frac{\partial^3 f(a,b)}{\partial x^3} (x-a)^3 + \frac{\partial^3 f(a,b)}{\partial y^3} (y-b)^3 + 3 \frac{\partial^3 f(a,b)}{\partial x^2 \partial y} (x-a)^2 (y-b) + 3 \frac{\partial^3 f(a,b)}{\partial x \partial y^2} (x-a) (y-b)^2] + \dots
$$

Hacemos la siguiente sustitución de variables: $x = S_t, \ y = t, \ a = S_{t_1}, \ b = t_1$

$$
f(S_t,t) = f(S_{t_1},t_1) + \frac{1}{1!} [\frac{\partial f(S_{t_1},t_1)}{\partial S_t} (S_t-S_{t_1}) + \frac{\partial f(S_{t_1},t_1)}{\partial t} (t-t_1)] + \\

\frac{1}{2!} [\frac{\partial^2 f(S_{t_1},t_1)}{\partial S_t^2} (S_t-S_{t_1})^2 + \frac{\partial^2 f(S_{t_1},t_1)}{\partial t^2} (t-t_1)^2 + 2 \frac{\partial^2 f(S_{t_1},t_1)}{\partial S_t \partial t} (S_t-S_{t_1})(t-t_1)] + \\

\frac{1}{3!} [\frac{\partial^3 f(S_{t_1},t_1)}{\partial S_t^3} (S_t-S_{t_1})^3 + \frac{\partial^3 f(S_{t_1},t_1)}{\partial t^3} (t-t_1)^3 + 3 \frac{\partial^3 f(S_{t_1},t_1)}{\partial S_t^2 \partial t} (S_t-S_{t_1})^2 (t-t_1) + 3 \frac{\partial^3 f(S_{t_1},t_1)}{\partial S_t \partial t^2} (S_t-S_{t_1}) (t-t_1)^2] + \dots 
$$

Consideremos lo siguiente:

(1):

$
S_{t_1} \rightarrow S_t \implies S_t - S_{t_1} = \Delta S_t
$

$
t_1 \rightarrow t \implies t - t_1 = \Delta t
$

$
f(S_t, t) - f(\Delta S_t, \Delta t) = \Delta f(S_t , t)
$

(2):

$
\Delta t \rightarrow 0 \implies \Delta t = dt;\ \  \Delta S_t = dS_t;\ \ \Delta f(S_t, t) = df(S_t, t)
$

Utilizando la propiedad (1) y (2) tenemos que:

$$
df(S_t,t) = \frac{1}{1!} [\frac{\partial f(S_t, t)}{\partial S_t} dS_t + \frac{\partial f(S_t,t)}{\partial t} dt] + \\

\frac{1}{2!} [\frac{\partial^2 f(S_t,t)}{\partial S_t^2} dS_t^2 + \frac{\partial^2 f(S_t,t)}{\partial t^2} dt^2 + 2 \frac{\partial^2 f(S_t,t)}{\partial S_t \partial t} dS_tdt] + \\

\frac{1}{3!} [\frac{\partial^3 f(S_t,t)}{\partial S_t^3} dS_t^3 + \frac{\partial^3 f(S_t,t)}{\partial t^3} dt^3 + 3 \frac{\partial^3 f(S_t,t)}{\partial S_t^2 \partial t} dS_t^2 dt + 3 \frac{\partial^3 f(S_t,t)}{\partial S_t \partial t^2} dS_t dt^2] + \dots 
$$

Recordando la multiplicación de diferenciales del cálculo de Itô:

<div align="center">

| Producto | $dt$ | $dW_t$ |
|----------|------|--------|
| $dt$     | $0$    | $0$      |
| $dW_t$   | $0$    | $dt$   |

</div>

Además que sabemos que $dS_t = S_t \sigma dW_t$, por lo tanto:

$$
d_St^2 = S_t^2 \sigma^2 dt \\
dS_t^3 = (S_t^2 \sigma^2 dt) (S_t \sigma dW_t) = 0 \\
dS_t^4 = 0 (S_t \sigma dW_t) = 0 \\
dS_t^5 = 0 (S_t \sigma dW_t) = 0 \\
\dots
$$

Entonces a partir de $dS_t^3$ en adelante se repite el patrón mostrado y se eliminan ya que por Itô se hacen cero esos diferenciales, además por Itô tenemos que $dS_t^2 dt = 0$ y $dS_t dt^2 = 0$. Aplicando estas propiedades eliminamos términos y llegamos a:

$$
df(S_t,t) = \frac{\partial f(S_t, t)}{\partial S_t} dS_t + \frac{\partial f(S_t, t)}{\partial t} dt + \frac{1}{2} \frac{\partial^2 f(S_t, t)}{dS_t^2} S_t^2 \sigma^2 dt
$$

Establecemos que $f(s_t,t) = f_t$, de forma final se tiene:

$$
\boxed{
df_t = \frac{\partial f_t}{\partial S_t} dS_t + \frac{\partial f_t}{\partial t} dt + \frac{1}{2} \frac{\partial^2 f_t}{\partial S_t^2} S_t^2 \sigma^2 dt
}
$$

## <font color='cornflowerblue'> Problema 3

Considere el proceso de difusión $dS_t = \mu S_t^3 + \sigma^2 dW_t + \frac{1}{2} dt$. Encuentre una forma diferencial de la función $f(S_{(t + \triangle t)}, t + \triangle t) = df_{(t + \triangle t)} := df_t$:


Tomando el modelo de expansión de Taylor para $k=3$:

$$

T(x, y) = f(a, b) + \frac{1}{1!} \left[\frac{\partial f(a, b)}{\partial x} (x-a) + \frac{\partial f(a, b)}{\partial y} (y-b)\right] + \\

\frac{1}{2!} \left[\frac{\partial^2 f(a,b)}{\partial x^2} (x-a)^2 + \frac{\partial^2 f(a,b)}{\partial y^2} (y-b)^2 + 2 \frac{\partial^2 f(a,b)}{\partial x \partial y} (x-a)(y-b) \right] + \\

\frac{1}{3!} \left[\frac{\partial^3 f(a,b)}{\partial x^3} (x-a)^3 + \frac{\partial^3 f(a,b)}{\partial y^3} (y-b)^3 + 3 \frac{\partial^3 f(a,b)}{\partial x^2 \partial y} (x-a)^2(y-b) + 3 \frac{\partial^3 f(a,b)}{\partial x \partial y^2} (x-a)(y-b)^2 \right] + ...

$$

Definiendo los cambios de variable $x=S_t$, $a=S_{t_1}$, $y=t$, $b=t_1$ es posible remplazar en la expansión de Taylor:

$$

f(S_t, t) = f(S_{t_1}, t_1) + \frac{1}{1!} \left[\frac{\partial f(S_{t_1}, t_1)}{\partial S_t} (S_t - S_{t_1}) + \frac{\partial f(S_{t_1}, t_1)}{\partial t} (t-t_1)\right] + \\

\frac{1}{2!} \left[\frac{\partial^2 f(S_{t_1}, t_1)}{\partial S_t^2} (S_t - S_{t_1})^2 + \frac{\partial^2 f(S_{t_1}, t_1)}{\partial t^2} (t-t_1)^2 + 2 \frac{\partial^2 f(S_{t_1}, t_1)}{\partial S_t \partial t} (S_t - S_{t_1})(t-t_1) \right] + \\

\frac{1}{3!} \left[\frac{\partial^3 f(S_{t_1}, t_1)}{\partial S_t^3} (S_t - S_{t_1})^3 + \frac{\partial^3 f(S_{t_1}, t_1)}{\partial t^3} (t-t_1)^3 + 3 \frac{\partial^3 f(S_{t_1}, t_1)}{\partial S_t^2 \partial t} (S_t - S_{t_1})^2(t-t_1) + 3 \frac{\partial^3 f(S_{t_1}, t_1)}{\partial S_t \partial t^2} (S_t-S_{t_1})(t-t_1)^2 \right] + ...

$$

Consideremos lo siguiente:

(1):

$
S_{t_1} \rightarrow S_t \implies S_t - S_{t_1} = \Delta S_t
$

$
t_1 \rightarrow t \implies t - t_1 = \Delta t
$

$
f(S_t, t) - f(\Delta S_t, \Delta t) = \Delta f(S_t , t)
$

(2):

$
\Delta t \rightarrow 0 \implies \Delta t = dt;\ \  \Delta S_t = dS_t;\ \ \Delta f(S_t, t) = df(S_t, t)
$

Utilizando (1) y (2) podemos reescribir de la siguiente manera:

$$

df(S_t, t) =  \frac{1}{1!} \left[\frac{\partial f(S_t, t)}{\partial S_t} dS_t + \frac{\partial f(S_t, t)}{\partial t} dt\right] + \\

\frac{1}{2!} \left[\frac{\partial^2 f(S_t, t)}{\partial S_t^2} dS_t^2 + \frac{\partial^2 f(S_t, t)}{\partial t^2} dt^2 + 2 \frac{\partial^2 f(S_t, t)}{\partial S_t \partial t} dS_tdt \right] + \\

\frac{1}{3!} \left[\frac{\partial^3 f(S_t, t)}{\partial S_t^3} dS_t^3 + \frac{\partial^3 f(S_t, t)}{\partial t^3} dt^3 + 3 \frac{\partial^3 f(S_t, t)}{\partial S_t^2 \partial t} dS_t^2dt + 3 \frac{\partial^3 f(S_t, t)}{\partial S_t \partial t^2} dS_tdt^2 \right] + ...

$$

Recordando la multiplicación de diferenciales del cálculo de Itô:

<div align="center">

| Producto | $dt$ | $dW_t$ |
|----------|------|--------|
| $dt$     | $0$    | $0$      |
| $dW_t$   | $0$    | $dt$   |

</div>

Podemos reescribir eliminando aquellos términos con $dt^2$, $dt^3$ y $dt\cdot dW_t$:

$$

df(S_t, t) =  \frac{1}{1!} \left[\frac{\partial f(S_t, t)}{\partial S_t} dS_t + \frac{\partial f(S_t, t)}{\partial t} dt\right] + 

\frac{1}{2!} \left[\frac{\partial^2 f(S_t, t)}{\partial S_t^2} dS_t^2 + 2 \frac{\partial^2 f(S_t, t)}{\partial S_t \partial t} dS_tdt \right] + 

\frac{1}{3!} \left[\frac{\partial^3 f(S_t, t)}{\partial S_t^3} dS_t^3 + 3 \frac{\partial^3 f(S_t, t)}{\partial S_t^2 \partial t} dS_t^2dt \right] + ...

$$

Dado que no conocemos el resultado de $dS_t^2$, $dS_t^3$, $dS_t \cdot dt$ y $dS_t \cdot dt$, habría que desarrollar $dS_t$ con la definición dada en el problema $dS_t = \mu S_t^3 + \sigma^2 dW_t + \frac{1}{2} dt$, y con ello, conocer si dichos términos se deben eliminar o deben permanecer:

Desarrollo para $dS_t^2$:

$$
dS_t^2 = \left(\mu S_t^3 + \sigma^2 dW_t + \frac{1}{2}dt \right)^2 = \left(\mu S_t^3 + \sigma^2 dW_t + \frac{1}{2}dt \right) \cdot \left(\mu S_t^3 + \sigma^2 dW_t + \frac{1}{2}dt \right)
$$

$$
dS_t^2 = \mu^2 S_t^6 + \mu S_t^3 \sigma^2 dW_t + \frac{1}{2}\mu S_t^3 dt + \mu S_t^3 \sigma^2 dW_t + \sigma^4 dW_t^2 + \frac{1}{2}\sigma^2 dt dW_t + \frac{1}{2}\mu S_t^3 dt + \frac{1}{2}\sigma^2 dt dW_t + \frac{1}{4} dt^2
$$

Eliminar $dt^2$, y $dt\cdot dW_t$ y reemplazar $dW_t^2 = dt$:

$$
dS_t^2 = \mu^2 S_t^6 + 2\mu S_t^3 \sigma^2 dW_t + \mu S_t^3 dt + \sigma^4 dt
$$

Desarrollo para $dS_t^3$:

$$
dS_t^3 = \left(dS_t^2 \right)(dS_t) = \left(\mu S_t^3 + \sigma^2 dW_t + \frac{1}{2}dt \right) \left(\mu^2 S_t^6 + 2\mu S_t^3 \sigma^2 dW_t + \mu S_t^3 dt + \sigma^4 dt \right)
$$

$$
dS_t^3 = \mu^3 S_t^9 + 2\mu^2 S_t^5 \sigma^2 dW_t + \mu^2 S_t^6 dt + \mu S_t^3 \sigma^4 dt + 

\mu^2 \sigma^2 S_t^6 dW_t + 2\mu \sigma^4 dW_t^2 + \mu \sigma^2 S_t^3 dt dW_t + \sigma^6 dW_t dt + 

\frac{1}{2} \mu^2 S_t^6 dt + \mu S_t^3 \sigma^2 dt dW_t + \frac{1}{2} \mu S_t^3 dt^2 + \frac{1}{2} \sigma^4 dt dW_t 
$$

Eliminar $dt^2$, y $dt\cdot dW_t$ y reemplazar $dW_t^2 = dt$:

$$
dS_t^3 = S_t^9 \mu^3  + 2 S_t^6 \mu^2 \sigma^2 dW_t + S_t^6 \mu^2 dt + S_t^3 \mu \sigma^4 dt + 

S_t^6 \mu^2 \sigma^2  dW_t + 2\mu \sigma^4 dt + 

\frac{1}{2} S_t^6 \mu^2  dt
$$

$$
dS_t^3 = S_t^9 \mu^3  + 3 S_t^6 \mu^2 \sigma^2 dW_t + S_t^6 \mu^2 dt + S_t^3 \mu \sigma^4 dt +  2\mu \sigma^4 dt + 

\frac{1}{2} S_t^6 \mu^2  dt
$$

Desarrollo para $dS_t \cdot dt$:

$$
dS_t \cdot dt = \left(\mu S_t^3 + \sigma^2 dW_t + \frac{1}{2} dt \right)(dt) \\
$$

$$
dS_t \cdot dt = \mu S_t^3 dt + \sigma^2 dt dW_t + \frac{1}{2} dt^2
$$

Eliminar $dt^2$, y $dt\cdot dW_t$:

$$
dS_t \cdot dt = \mu S_t^3 dt 
$$

Desarrollo para $dS_t^2 \cdot dt$:

$$
dS_t^2 \cdot dt = \left(dS_t \cdot dt \right) (dS_t) = \left(\mu S_t^3 dt \right) \left(\mu S_t^3 + \sigma^2 dW_t + \frac{1}{2} dt \right)
$$

$$
dS_t^2 \cdot dt = \mu^2 S_t^6 dt + \mu S_t^3 \sigma^2 dt dW_t + \frac{1}{2} \mu S_t^3 dt^2
$$

Eliminar $dt^2$, y $dt\cdot dW_t$:

$$
dS_t^2 \cdot dt = \mu^2 S_t^6 dt
$$

Con lo anterior se ha demostrado que ninguno de los términos $dS_t^2$, $dS_t^3$, $dS_t \cdot dt$ y $dS_t \cdot dt$ es igual a cero, por lo que deben permanecer en el modelo y no deberían eliminarse. Por lo tanto, el modelo sería:

$$

df(S_t, t) = \left[\frac{\partial f(S_t, t)}{\partial S_t} dS_t + \frac{\partial f(S_t, t)}{\partial t} dt\right] + 

\frac{1}{2} \left[\frac{\partial^2 f(S_t, t)}{\partial S_t^2} dS_t^2 + 2 \frac{\partial^2 f(S_t, t)}{\partial S_t \partial t} dS_tdt \right] + 

\frac{1}{6} \left[\frac{\partial^3 f(S_t, t)}{\partial S_t^3} dS_t^3 + 3 \frac{\partial^3 f(S_t, t)}{\partial S_t^2 \partial t} dS_t^2dt \right] + ...

$$

Definiendo $f(S_t, t) = f_t$, de forma final se tiene:

$$
\boxed{
df_t = \left[\frac{\partial f_t}{\partial S_t} dS_t + \frac{\partial f_t}{\partial t} dt\right] + 

\frac{1}{2} \left[\frac{\partial^2 f_t}{\partial S_t^2} dS_t^2 + 2 \frac{\partial^2 f_t}{\partial S_t \partial t} dS_tdt \right] + 

\frac{1}{6} \left[\frac{\partial^3 f_t}{\partial S_t^3} dS_t^3 + 3 \frac{\partial^3 f_t}{\partial S_t^2 \partial t} dS_t^2dt \right] + ...
}
$$
