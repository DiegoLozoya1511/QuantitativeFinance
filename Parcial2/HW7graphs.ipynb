```python
import numpy as np
import matplotlib.pyplot as plt

# ========== 1. Binary Put Cash-or-Nothing ==========
def plot_put_cash_or_nothing(K=1.0, c=1.0, x_max=2.0):
    x = np.linspace(0, x_max, 400)

    plt.figure(figsize=(5,4), dpi=150)

    # Left segment (S_T <= K): payoff = c
    plt.hlines(c, 0, K, color='cornflowerblue', linewidth=2)

    # Right segment (S_T > K): payoff = 0
    plt.hlines(0, K, x_max, color='cornflowerblue', linewidth=2)

    # Markers at K
    plt.scatter([K], [c], color='cornflowerblue', s=50)  # closed circle (<= branch)
    plt.scatter([K], [0], facecolors='none', edgecolors='cornflowerblue', s=50)  # open circle (> branch)

    # Labels
    plt.text(K, -0.25*c, r'$K$', ha='center', va='top')
    plt.text(-0.025*x_max, c, r'$c$', va='center', ha='right')

    # Add "0" label at left
    plt.text(-0.025*x_max, 0, r'$0$', va='center', ha='right')

    plt.tick_params(labelbottom=False, labelleft=False)  # hides numbers, keeps ticks (and grid)

    # Formatting
    plt.title('Binary Put Cash-or-Nothing')
    plt.grid(alpha=0.3)
    plt.ylim(-0.2*c, 1.2*c)
    plt.xlim(0, x_max)
    plt.show()

# Example call
plot_put_cash_or_nothing()
```

```python
# ========== 2. Binary Call Cash-or-Nothing ==========
def plot_call_cash_or_nothing(K=1.0, c=1.0, x_max=2.0):
    plt.figure(figsize=(5,4), dpi=150)

    # Left segment (S_T <= K): payoff = 0
    plt.hlines(0, 0, K, color='cornflowerblue', linewidth=2)

    # Right segment (S_T > K): payoff = c
    plt.hlines(c, K, x_max, color='cornflowerblue', linewidth=2)

    # Markers at K
    plt.scatter([K], [0], color='cornflowerblue', s=50)  # closed circle (<= branch)
    plt.scatter([K], [c], facecolors='none', edgecolors='cornflowerblue', s=50)  # open circle (> branch)

    # Labels
    plt.text(K, -0.25*c, r'$K$', ha='center', va='top')
    plt.text(-0.025*x_max, c, r'$c$', va='center', ha='right')

    # Add "0" label at left
    plt.text(-0.025*x_max, 0, r'$0$', va='center', ha='right')

    # Hide tick labels but keep grid
    plt.tick_params(labelbottom=False, labelleft=False)
    plt.grid(alpha=0.3)

    # Formatting
    plt.title('Binary Call Cash-or-Nothing')
    plt.ylim(-0.2*c, 1.2*c)
    plt.xlim(0, x_max)
    plt.show()

# Example call
plot_call_cash_or_nothing()
```

```python
# ========== 3. Binary Put Asset-or-Nothing ==========
def plot_put_asset_or_nothing(K=1.0, x_max=2.0):
    plt.figure(figsize=(5,4), dpi=150)

    # Left segment (S_T <= K): payoff = S_T (line with slope 1)
    plt.plot([0, K], [0, K], color='cornflowerblue', linewidth=2)

    # Right segment (S_T > K): payoff = 0
    plt.hlines(0, K, x_max, color='cornflowerblue', linewidth=2)

    # Markers at K
    plt.scatter([K], [K], color='cornflowerblue', s=50)  # open circle (< branch)
    plt.scatter([K], [0], facecolors='none', edgecolors='cornflowerblue', s=50)  # closed circle (>= branch)

    # Labels
    plt.text(K, -0.25*K, r'$K$', ha='center', va='top')
    plt.text(-0.025*x_max, K, r'$K$', va='center', ha='right')
    plt.text(-0.025*x_max, 0, r'$0$', va='center', ha='right')

    # Hide tick labels but keep grid
    plt.tick_params(labelbottom=False, labelleft=False)
    plt.grid(alpha=0.3)

    # Formatting
    plt.title('Binary Put Asset-or-Nothing')
    plt.ylim(-0.2*K, 1.2*K)
    plt.xlim(0, x_max)
    plt.show()

# Example call
plot_put_asset_or_nothing()
```

```python
# ========== 4. Binary Call Asset-or-Nothing ==========
def plot_call_asset_or_nothing(K=1.0, x_max=2.0):
    plt.figure(figsize=(5,4), dpi=150)

    # Left segment (S_T <= K): payoff = 0
    plt.hlines(0, 0, K, color='cornflowerblue', linewidth=2)

    # Right segment (S_T > K): payoff = S_T (line with slope 1)
    plt.plot([K, x_max], [K, x_max], color='cornflowerblue', linewidth=2)

    # Markers at K
    plt.scatter([K], [0], color='cornflowerblue', s=50)  # closed circle (<= branch)
    plt.scatter([K], [K], facecolors='none', edgecolors='cornflowerblue', s=50)  # open circle (> branch)

    # Labels
    plt.text(K, -0.25*K, r'$K$', ha='center', va='top')
    plt.text(-0.025*x_max, K, r'$K$', va='center', ha='right')
    plt.text(-0.025*x_max, 0, r'$0$', va='center', ha='right')

    # Hide tick labels but keep grid
    plt.tick_params(labelbottom=False, labelleft=False)
    plt.grid(alpha=0.3)

    # Formatting
    plt.title('Binary Call Asset-or-Nothing')
    plt.ylim(-0.2*K, 1.5*K)
    plt.xlim(0, x_max)
    plt.show()

# Example call
plot_call_asset_or_nothing()
```
